```python
import pandas as pd
```

```python
schools_df = "../Resources/schools_complete.csv"
students_df = "../Resources/students_complete.csv"
```

```python
schools_data_df = pd.read_csv(schools_df)
schools_data_df.head()
```

```python
students_data_df = pd.read_csv(students_df)
students_data_df.head()
```

```python
schools_combined = pd.merge (students_data_df, schools_data_df, how="left", on=["school_name", "school_name"])
schools_combined.head()
```

```python
#Total # of schools
total_number_schools=len(schools_data_df["School ID"].unique())
total_number_schools
                         
```

```python
#Total # of students
total_number_students = len(students_data_df ["Student ID"].unique())
total_number_students
```

```python
#Total budget
total_budget = schools_data_df["budget"].sum()
total_budget
```

```python
#Average math score
avg_math_score = students_data_df["math_score"].mean()
avg_math_score
```

```python
#Average reading score
avg_read_score = students_data_df["reading_score"].mean()
avg_read_score
```

```python
# of students passing math (70% or greater)
students_passing_math = students_data_df.loc[students_data_df ["math_score"] >=70, "math_score"]
students_passing_math.count()
```

```python
# % of passing math score (70% or greater)
percent_passing_math = float(students_passing_math.count()/total_number_students)*100
percent_passing_math
```

```python
# of students passing reading (70% or greater)
students_passing_reading = students_data_df.loc[students_data_df ["reading_score"] >=70, "reading_score"]
students_passing_reading.count()
```

```python
# % of passing reading score (70% or greater)
percent_passing_reading = float(students_passing_reading.count()/total_number_students)*100
percent_passing_reading
```

```python
# % of students who passed math and reading (% Overall Passing)
overall_math_reading = students_data_df[(students_data_df["math_score"] >= 70) & (students_data_df["reading_score"] >= 70)]
overall_math_reading["student_name"].count()

overall_passing_count = overall_math_reading["student_name"].count()

percent_overall_passing = (overall_passing_count/total_number_students)*100
percent_overall_passing
```

```python
#Create dataframe to hold above results
schools_summary_df = pd.DataFrame(schools_combined)
```

```python
#School Summary
school_type = schools_combined.groupby("school_name")["type"].unique()
school_type

```

```python
total_students = schools_combined["school_name"].value_counts()
total_students
```

```python
school_budget = schools_combined.groupby("school_name")["budget"].mean()
school_budget
```

```python
per_student_budget = school_budget/total_students
per_student_budget
```

```python
avg_math_score = schools_combined.groupby("school_name")["math_score"].mean()
avg_math_score
```

```python
avg_reading_score = schools_combined.groupby("school_name")["reading_score"].mean()
avg_reading_score
```

```python
passing_math = schools_combined[schools_combined["math_score"] >= 70].groupby("school_name")["Student ID"].count()
percent_passing_math = (passing_math/total_students)*100
percent_passing_math
```

```python
passing_reading = schools_combined[schools_combined["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
percent_passing_reading = (passing_reading/total_students)*100
percent_passing_reading
```

```python
per_passing_math_reading = schools_combined[(schools_combined["math_score"] >= 70) & (schools_combined["reading_score"] >= 70)]
per_passing_math_reading = per_passing_math_reading.groupby(["school_name"]).count()["student_name"]
per_passing_math_reading
```

```python
percent_overall_passing = (per_passing_math_reading / total_students) * 100
percent_overall_passing
```

```python
#School Summary DataFrame
schools_summary_df = pd.DataFrame (
        {"School Type" : school_type,
         "Total Students" : total_students,
         "Total School Budget" : school_budget,
         "Per Student Budget" : per_student_budget, 
         "Average Math Score" : round(avg_math_score,2),
         "Average Reading Score" : round(avg_reading_score,2),
         "% Passing Math" : round(percent_passing_math,2),
         "% Passing Reading" : round(percent_passing_reading,2),
         "% Overall Passing" : round(percent_overall_passing,2),})
```

```python
#Table
schools_summary_df["Total Students"] = schools_summary_df["Total Students"].map("{:,.0f}".format)
schools_summary_df["Total School Budget"] = schools_summary_df["Total School Budget"].map("${:,.2f}".format)
schools_summary_df["Per Student Budget"] = schools_summary_df["Per Student Budget"].map("${:,.2f}".format)

schools_summary_df
```

```python
#Top 5 % Overall Passing
top_five_schools = schools_summary_df.sort_values(["% Overall Passing"], ascending=False)
top_five_schools.head(5)
```

```python
#Bottom 5 Lowest Performaing Schools by % 
lowest_five_schools = schools_summary_df.sort_values(["% Overall Passing"], ascending=True)
lowest_five_schools.head(5)
```

```python
#Math Scores by Grade
ninth_grade_math =  schools_combined.loc[schools_combined["grade"] == "9th"].groupby("school_name")["math_score"].mean()
tenth_grade_math =  schools_combined.loc[schools_combined["grade"] == "10th"].groupby("school_name")["math_score"].mean()
eleventh_grade_math =  schools_combined[schools_combined["grade"] == "11th"].groupby("school_name")["math_score"].mean()
twelveth_grade_math =  schools_combined[schools_combined["grade"] == "12th"].groupby("school_name")["math_score"].mean()

#DataFrame info 
avg_math_by_school_df = pd.DataFrame(
    {"9th" : ninth_grade_math, 
     "10th" : tenth_grade_math, 
     "11th" : eleventh_grade_math,
     "12th": twelveth_grade_math,})           

avg_math_by_school_df
```

```python
#Reading Scores by Grade
nineth_grade_reading =  schools_combined.loc[schools_combined["grade"] == "9th"].groupby("school_name")["reading_score"].mean()
tenth_grade_reading =  schools_combined.loc[schools_combined["grade"] == "10th"].groupby("school_name")["reading_score"].mean()
eleventh_grade_reading =  schools_combined.loc[schools_combined["grade"] == "11th"].groupby("school_name")["reading_score"].mean()
twelveth_grade_reading =  schools_combined.loc[schools_combined["grade"] == "12th"].groupby("school_name")["reading_score"].mean()

#DataFrame info
avg_reading_by_school_df = pd.DataFrame(
    {"9th":nineth_grade_reading,
     "10th":tenth_grade_reading,
     "11th":eleventh_grade_reading, 
     "12th":twelveth_grade_reading,})            

avg_reading_by_school_df
```

```python
#Scores by School Spending
spending_bins = [0, 585, 630, 645, 700]
group_names = [" < $585", "$581-630", "$631-645", " > $645"]


schools_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)
schools_summary_df.head()


school_spending_grouped = schools_summary_df.groupby("Spending Ranges (Per Student)").mean() 

school_spending_grouped
```

```python
#Scores by School Size
size_bins = [0, 1000, 2000, 5000]
bins_names = ["Small ( <1000)", "Medium (1000-2000)", "Large (2000-5000)"]


schools_summary_df["School Size"] = pd.cut(total_students, size_bins, labels = bins_names)
schools_summary_df.head()

school_size_grouped = schools_summary_df.groupby("School Size").mean() 

school_size_grouped
```

```python
#Scores by School Type
scores_by_school_type = schools_summary_df
scores_by_school_type["School Type"] = scores_by_school_type["School Type"].astype(str)
scores_by_school_type.groupby("School Type").mean()
```
